```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import pandas as pd
```

```python
def kernel_density(x: np.ndarray, h: float) -> None:
  t = np.linspace(min(x) - h, max(x) + h,num=len(x))
  j = np.array([])
  for i in range(t.size):
    val = J(t[i],x, h)
    j = np.append(j,val)
  plt.plot(t,j)
def J(t,X,h):
  n=len(X)
  j = 0
  for x in X:
    j += cosine_kernel((x-t) / h)
  return j / (n * h)
def cosine_kernel(u):
  return 1 + m.cos(2*m.pi*u) if abs(u) <= 0.5 else 0 
```

```python
grades = pd.DataFrame(pd.read_csv("grades.csv"))
kernel_density(grades["math"],12)
```

```python
h = [i for i in range(1, 37)]
for val in h:
  kernel_density(grades["math"],val)
```

```python
h=13
x1=grades[grades["school"] == 'A']["math"]
x2=grades[grades["school"] == 'B']["math"]
x3=grades[grades["school"] == 'C']["math"]
kernel_density(x1,h)
kernel_density(x2,h)
kernel_density(x3,h)
plt.title("Math grades by school")
plt.legend(['A','B','C'])
plt.show()
```
