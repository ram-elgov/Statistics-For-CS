```python
import math as m

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def kernel_density(x_vec: np.ndarray, h: float) -> None:
    t = np.linspace(min(x_vec) - h, max(x_vec) + h, num=len(x_vec))
    j = np.array([J(sample, x_vec, h) for sample in t])
    plt.plot(t, j)

def J(a: float, x_vec: np.ndarray, h: float):
    n = len(x_vec)
    j = sum(cosine_kernel((x - a) / h) for x in x_vec)
    return j / (n * h)

def cosine_kernel(u: float) -> float:
    return 1 + m.cos(2 * m.pi * u) if abs(u) <= 0.5 else 0.0
```

```python
grades = pd.DataFrame(pd.read_csv("grades.csv"))
kernel_density(grades["math"],12)
```

```python
h = range(1, 37)
for width in h:
  kernel_density(grades["math"],width)
```

```python
h = 13
A = grades[grades["school"] == "A"]["math"]
B = grades[grades["school"] == "B"]["math"]
C = grades[grades["school"] == "C"]["math"]
kernel_density(A, h)
kernel_density(B, h)
kernel_density(C, h)
plt.title("Math grades by school")
plt.legend(["A", "B", "C"])
plt.show()
```
