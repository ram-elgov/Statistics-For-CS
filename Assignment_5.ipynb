# Problem 1

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

rv = norm(loc=81.5, scale=13)
rv_50 = norm(loc=81.5, scale = 13 / 50 ** 0.5)
rv_25 = norm(loc=81.5, scale = 13 / 25 ** 0.5)
print(f"Probability to weigh more than 82kg: {rv.sf(82):.3f}")
print(f"Probability for 50 plants to weigh more than 82kg: {rv.sf(82) ** 50}")
print(f"Probabilty for 50 plants to average more than 82kg: {rv_50.sf(82):.3f}")
print(f"Probabilty for 25 plants to weigh between 80-82kg: {(rv_25.cdf(82) - rv_25.cdf(80)):.3f}")
print(f"10% of the plants weigh more than: {rv.ppf(0.9):.3f}")
```

# Problem 2

```python
from scipy.stats import randint

X = randint(1, 7)
for val in range(X.a, X.b + 1):
  print(f"P({val}) = {X.pmf(val):.3f}")

print()
print(f"Mean: {X.mean():.3f}")
print(f"Std: {X.std():.3f}")
```

```python
from seaborn import kdeplot

rolls = 50000
dice_count = 2
avg_roll_vec = sum(X.rvs(rolls) for _ in range(dice_count)) / dice_count

kdeplot(avg_roll_vec)
plt.show()
```

```python
empirical_mean, empirical_std = avg_roll_vec.mean(), avg_roll_vec.std()
print(f"Empirical mean: {empirical_mean}")
print(f"Empirical std: {empirical_std}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

X2_norm = norm(loc=empirical_mean, scale=empirical_std)

kdeplot(avg_roll_vec)

x = np.linspace(0, 6.5, 5000)
y = X2_norm.pdf(x)
plt.plot(x, y)
```

```python
from scipy.stats import uniform

X = uniform(0, 7)
rolls = 50000
dice_count = 2
avg_roll_vec = sum(X.rvs(rolls) for _ in range(dice_count)) / dice_count

kdeplot(avg_roll_vec)
plt.show()
```

```python
empirical_mean, empirical_std = avg_roll_vec.mean(), avg_roll_vec.std()
print(f"Empirical mean: {empirical_mean:.3f}")
print(f"Empirical std: {empirical_std:.3f}")
```

```python
X2_norm = norm(loc=empirical_mean, scale=empirical_std)

kdeplot(avg_roll_vec)

x = np.linspace(0, 6.5, 5000)
y = X2_norm.pdf(x)
plt.plot(x, y)
plt.show()
```

```python
rolls = 50000
for dice_count in range(2, 50, 10):
  avg_roll_vec = sum(X.rvs(rolls) for _ in range(dice_count)) / dice_count
  kdeplot(avg_roll_vec)
  empirical_mean, empirical_std = avg_roll_vec.mean(), avg_roll_vec.std()
  print(f"Summary for {dice_count} dice: ")
  print(f"Empirical mean: {empirical_mean:.3f}")
  print(f"Empirical std: {empirical_std:.3f}")
  print()
```

# Problem 4

```python
# MSE(mu) graph for 50<=mu<=70 y axis is restricted to MSE<=10
x = np.linspace(50, 70, 100)
mse1 = (60 - x) ** 2
mse2 = x * 0 + 2.5
mse3 = (((x/6) - 10) ** 2) + (125 / 72)
plt.plot(x, mse1, 'r', label='mse1', color="blue")
plt.plot(x, mse2, 'r', label='mse2', color="red")
plt.plot(x, mse3, 'r', label='mse3', color="green")
plt.ylim(0,10)
idx = np.argwhere(np.diff(np.sign(mse3 - mse2))).flatten()
plt.plot(x[idx], mse2[idx], 'ro')
plt.xlabel("mu")
plt.ylabel("MSE(mu)")
plt.title("MSE comparison")
plt.legend()
plt.show()
print(f"Intersection points of mse2 and mse3: {x[idx]}")
```
